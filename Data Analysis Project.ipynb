```python
#conver into date time
df["reservation_status_date"]=pd.to_datetime(df["reservation_status_date"],format="%d/%m/%Y")
```

```python
#data is 1lakh+ and company & agent also 1 llakh+ so remove them and drop null values 
df.drop(["company","agent"],axis=1,inplace=True)
#drop null values from every column which contain null
df.dropna(inplace=True)
```

```python
df=df[df["adr"]<5000]
```

# Data Analysys and visualization

```python
cancelled_per=df["is_canceled"].value_counts(normalize=True)
print(cancelled_per)
plt.figure(figsize=(5,4))
plt.title("Reservation status")
plt.bar(["Not Cancled","Cancled"],df["is_canceled"].value_counts(),color="pink")
plt.show()
```

```python
ax1 = sns.countplot(x="hotel", hue="is_canceled", data=df, palette="pink")
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(legend_labels, ["Not Cancel", "Cancel"], bbox_to_anchor=(1, 1))
plt.title("Reservation Status in different hotels")
plt.xlabel("Hotel")
plt.ylabel("No of Reservations")
plt.show()
```

```python
#cancelation ratio:
Resort_hotel=df[df["hotel"]=="Resort Hotel"]
Resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
#cancelation ratio:
City_hotel=df[df["hotel"]=="City Hotel"]
City_hotel["is_canceled"].value_counts(normalize=True)
```

```python
Resort_hotel=Resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
City_hotel=City_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
#price may be the reasone
plt.figure(figsize=(20,8))
plt.title("Avarage Dailt Rte in City and Resort Hotel",fontsize=30)
plt.plot(Resort_hotel.index,Resort_hotel["adr"],label="Resort_hotel")
plt.plot(City_hotel.index,City_hotel["adr"],label="City_hotel")
plt.legend(fontsize=20)
plt.show()
```

#### Conclusion: We can say that the price in Resort Hotel is more compear to City hotel,and the spick may be the weeak end day price

```python
df["month"]=df["reservation_status_date"].dt.month
plt.figure(figsize=(16,8))
axl=sns.countplot(x="month",hue="is_canceled",data=df,palette="viridis")
#legend_labels,_=axl.get_leged_handles_labels()
legend_labels, _ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status per month",size=18)
plt.xlabel("month")
plt.ylabel("no of reservation")
plt.legend(["not cancle","cancle"])
plt.show()
```

#### price have any effect on cancelation

```python
plt.figure(figsize=(15,8))
plt.title("ADR per month",fontsize=30)
sns.barplot(x="month",y="adr",data=df[df["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index(),palette="viridis")
plt.show()
```

## onlie TA is more than that ofline Ta

```python

market_segment_counts = df['market_segment'].value_counts()
plt.pie(market_segment_counts, labels=None, autopct=None, startangle=90, wedgeprops=dict(width=0.3))

# Draw a white circle in the center to create the donut chart effect
centre_circle = plt.Circle((0, 0), 0.2, color='white', edgecolor='black', linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Place the legend outside the pie chart and adjust the position
plt.legend(labels=market_segment_counts.index, bbox_to_anchor=(1, 0.5), loc="center left", title='Market Segments', title_fontsize='12')

plt.title('Distribution of Market Segments (Donut Chart)')
plt.show()
```

```python
cancelled_data=df[df["is_canceled"]==1]
cancelled_data["market_segment"].value_counts(normalize=True)
```

#### Conclusion: So from this above explanation we can say that online TA is more and online Cancelation is aslo more,may be the Online Hotel description is not exactely same as Real condition of Hotel

```python
cancelled_df_adr=cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values("reservation_status_date",inplace=True)

not_cancelled_data=df[df["is_canceled"]==0]
not_cancelled_df_adr=not_cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values("reservation_status_date",inplace=True)

plt.figure(figsize=(18,6))
plt.title("Avarage Daily Rate",fontsize=20)
plt.plot(cancelled_df_adr["reservation_status_date"],cancelled_df_adr["adr"],label="Cancelled")
plt.plot(not_cancelled_df_adr["reservation_status_date"],not_cancelled_df_adr["adr"],label="Not Cancelled")
plt.legend()
```
